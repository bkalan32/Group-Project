```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import hvplot.pandas
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python

# read the data from a CSV file
eth_contract = pd.read_csv(r"C:\Users\bkala\Downloads\export-verified-contracts.csv")
eth_transactions = pd.read_csv(r"C:\Users\bkala\Downloads\export-EtherPrice.csv")

display(eth_contract)
display(eth_transactions)
```

```python
# Convert Date(UTC) column to datetime format
eth_contract['Date(UTC)'] = pd.to_datetime(eth_contract['Date(UTC)'])

# Extract the year from the Date(UTC) column and add it as a new column
eth_contract['Year'] = eth_contract['Date(UTC)'].dt.year

# Group the data by year and calculate the sum of No. of Verified Contracts for each year
contracts_groups = eth_contract.groupby('Year', as_index=False)['No. of Verified Contracts'].sum()

# Split the data into training and testing sets
train_data = contracts_groups[contracts_groups['Year'] <= 2022]
test_data = contracts_groups[contracts_groups['Year'] > 2022]

# Create a linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(train_data[['Year']], train_data['No. of Verified Contracts'])

# Predict the No. of Contracts for the next 5 years
future_years = pd.DataFrame({'Year': [2023, 2024, 2025, 2026, 2027]})
future_contracts = model.predict(future_years)

# Print the predicted No. of Contracts for the next 5 years
print(future_contracts)
```

```python
# convert the numpy array to a pandas DataFrame
df_future_contracts = pd.DataFrame({'Year': np.arange(2023, 2028), 'No. of Verified Contracts': future_contracts})

# plot the data using hvplot
df_future_contracts.hvplot(x='Year', y='No. of Verified Contracts', kind='line', title='Future Contracts',yformatter='%0f.')
```

```python
# convert the Date(UTC) column to a datetime object
eth_contract['Date(UTC)'] = pd.to_datetime(eth_contract['Date(UTC)'])

# extract the year from the Date(UTC) column
eth_contract['Year'] = eth_contract['Date(UTC)'].dt.year

# group the data by year and sum the No. of Verified Contracts column
contracts_group = eth_contract.groupby('Year')['No. of Verified Contracts'].sum()

# print the grouped data
print(contracts_group)
```

## Show and Graph Eth Transactions

```python
# Rename Column Value to transactions
eth_transactions = eth_transactions.rename(columns={'Value': 'Transactions',})

# convert the Date(UTC) column to a datetime object
eth_transactions['Date(UTC)'] = pd.to_datetime(eth_transactions['Date(UTC)'])

# extract the year from the Date(UTC) column
eth_transactions['Year'] = eth_transactions['Date(UTC)'].dt.year

# group the data by year and sum the No. of Verified Contracts column
transactions_data = eth_transactions.groupby('Year')['Transactions'].sum()

# print the grouped data
print(transactions_data)
```

```python
transactions_data.hvplot.line(x='Year', 
                              y='Transactions',
                             title='Ethereum Transaction Volume from 2015-2023',
                             color='Red',
                             xlabel='Year',
                             ylabel='Transactions',
                             yformatter='%0f.')
```

### Combine the 2 datasets and plot them

```python
eth_merged = pd.merge(eth_contract, eth_transactions[['Date(UTC)', 'Transactions']], on='Date(UTC)', how='outer')

eth_merged = eth_merged.drop('UnixTimeStamp', axis=1)

eth_merged
```

```python
# Groupby and Plot dataframe
eth_merged_group = eth_merged.groupby('Year').agg({'No. of Verified Contracts': 'sum', 'Transactions': 'sum'})


eth_merged_group.hvplot.line(xlabel='Year',ylabel='No. of Verified Contracts/Transactions',yformatter='%0f.')
```
